# Information Entropy -

The measure of disorder.

#### Range - (0 -> 1)

Does a split increase our understanding of the data? if yes it adds a split or else no
A split is also called a leaf.

We have to find out the needed "leaves" of the algorithm

#### Average level of information / uncertainty inherent to the variables outcomes.

No Feature Scaling Required.



#### Importing Libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing Datasets

```python
dataset = pd.read_csv("Position_Salaries.csv")
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

#### Training the model

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
```

#### Predicting Outcome

```python
regressor.predict([[6.5]])
```

#### Visualising the DTR

```python
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.plot(x,y, color="green")
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

As we can see it doesn't really give usable outcomes for this small dataset
this is mostly useful for 3-dim datasets and sets with more complex columns and data.
