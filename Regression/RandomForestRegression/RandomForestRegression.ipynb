# Random Forest Regression

1. Pick random data points from datasets
2. Build decision tree based on it
3. Choose number of trees you want to build and repeat steps
4. Make prediction from all trees and output the average amongst all trees


## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import Datasets

```python
dataset = pd.read_csv("Position_Salaries.csv")

x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Train the Data

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=1000, random_state=0)
regressor.fit(x,y)
```

## Predict Results

```python
regressor.predict([[6.5]])
```

## Visualising the data

```python
X_grid = np.arange(min(x), max(x), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='green')
# plt.plot(x,y, color="blue")
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
